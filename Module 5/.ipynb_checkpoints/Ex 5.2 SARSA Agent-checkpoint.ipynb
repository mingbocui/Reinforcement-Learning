```python
class Agent(object):  
        
    def __init__(self, actions):
        self.actions = actions
        self.num_actions = len(actions)

    def act(self, state):
        raise NotImplementedError
```

```python
class SarsaAgent(Agent):
    
    def __init__(self, actions, epsilon=0.01, alpha=0.5, gamma=1):
        super(SarsaAgent, self).__init__(actions)
        
        ## Initialize empty dictionary here
        ## In addition, initialize the value of epsilon, alpha and gamma
        self.Q = dict()
        self.epsilon = epsilon
        self.gamma = gamma
        self.alpha = alpha
        
    def stateToString(self, state):
        mystring = ""
        if np.isscalar(state):
            mystring = str(state)
        else:
            for digit in state:
                mystring += str(digit)
        return mystring    
    
    def act(self, state):
        stateStr = self.stateToString(state)      
        action = np.random.randint(0, self.num_actions) 

        ## Implement epsilon greedy policy here
        if np.random.random() < self.epsilon:
            action = action
#             print("here")
        else:
            q_values = [self.Q.get((stateStr, a), 0) for a in self.actions]
#             print(q_values)
            action_ind = np.argmax(q_values)
            action = self.actions[action_ind]
            
        return action

    def learn(self, state1, action1, reward, state2, action2):
        state1Str = self.stateToString(state1)
        state2Str = self.stateToString(state2)

        ## Implement the sarsa update here
        q_value = self.Q.get((state1Str, action1), 0)
        q_value_prime = self.Q.get((state2Str, action2), 0)
        self.Q[(state1Str, action1)] = q_value + self.alpha * (reward + self.gamma * q_value_prime - q_value)
        """
        SARSA Update
        Q(s,a) <- Q(s,a) + alpha * (reward + gamma * Q(s',a') - Q(s,a))
        or
        Q(s,a) <- Q(s,a) + alpha * (td_target - Q(s,a))
        or
        Q(s,a) <- Q(s,a) + alpha * td_delta
        """
```

```python
class SarsaAgent(Agent):

    def __init__(self, actions, epsilon=0.01, alpha=0.5, gamma=1):

        super(SarsaAgent, self).__init__(actions)
        ## TODO 1
        ## Initialize empty dictionary here
        ## In addition, initialize the value of epsilon, alpha and gamma
        ## My code -----------------------------------------------------------------##

        # Initialize an empty dictionary for state-action value function
        self.q = {}
        # Initialize the poliy as an empty dictionary
        self.policy = {}

        # Initialize parameters
        self.epsilon = epsilon
        self.alpha = alpha
        self.gamma = gamma

        ##--------------------------------------------------------------------------##

    def stateToString(self, state):
        mystring = ""
        if np.isscalar(state):
            mystring = str(state)
        else:
            for digit in state:
                mystring += str(digit)
        return mystring    

    def act(self, state):

        stateStr = self.stateToString(state)
        action = np.random.randint(0, self.num_actions) 
        ## TODO 2
        ## Implement epsilon greedy policy here

        ## My code -----------------------------------------------------------------##
        # check if th current state is present into the policy dictionary
        if stateStr in self.policy.keys():
            # choose action by using the corresponding epsilon-greedy policy
            action = np.random.choice(self.num_actions, p=self.policy[stateStr])
        # else create an epsilon-greedy policy for the current state
        else:
            #policy_current_state = np.ones(self.num_actions)*(self.epsilon/self.num_actions)
            #policy_current_state[action] += 1. - self.epsilon
            policy_current_state = np.ones(self.num_actions)/self.num_actions
            # update initialization of policy
            self.policy[stateStr] = policy_current_state
            # choose an action by the updated policy
            action = np.random.choice(self.num_actions, p=self.policy[stateStr])

        return action

    def learn(self, state1, action1, reward, state2, action2):

        state1Str = self.stateToString(state1)
        state2Str = self.stateToString(state2)
        ## Implement the sarsa update here
        ## My code -----------------------------------------------------------------##
        # assure that also state1Str and state2Str are presents in q
        # check if state1Str and state2Str as keys are presents in q
        list_state_keys_to_add = []
        if state1Str not in self.q.keys():
            list_state_keys_to_add.append(state1Str)

        if state2Str not in self.q.keys():
            list_state_keys_to_add.append(state2Str)

        if len(list_state_keys_to_add) > 0:
            action_value_dict = {}
            for a in range(self.num_actions):
                action_value_dict[a] = 0.
            for state_key in list_state_keys_to_add:
                self.q[state_key] = action_value_dict

        # update Q(s,a) 
        target = reward + self.gamma*self.q[state2Str][action2]
        delta = target - self.q[state1Str][action1]
        self.q[state1Str][action1] += self.alpha*delta
        # update policy
        best_actions = []
        max_value = -np.inf
        for action, value in self.q[state1Str].items():
            # check if the current value is more than the max
            if value > max_value:
                # clear the list best_actions if it's not empty
                if len(best_actions) > 0:
                    best_actions.clear()
                # append the current action in best_actions
                best_actions.append(action)
                # update max_value
                max_value = value
            elif value == max_value:
                # append the current action in best_actions
                best_actions.append(action)
        self.policy[state1Str] = np.ones(self.num_actions)*(self.epsilon/self.num_actions)

        for a in best_actions:
            self.policy[state1Str][a] += (1. - self.epsilon)/len(best_actions)

            

        

        ##--------------------------------------------------------------------------##

        """

        SARSA Update

        Q(s,a) <- Q(s,a) + alpha * (reward + gamma * Q(s',a') - Q(s,a))

        or

        Q(s,a) <- Q(s,a) + alpha * (td_target - Q(s,a))

        or

        Q(s,a) <- Q(s,a) + alpha * td_delta

        """
```
